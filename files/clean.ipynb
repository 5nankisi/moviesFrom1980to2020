# Import necessary libraries

```python
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical operations
import matplotlib.pyplot as plt  # For plotting and visualization
import seaborn as sns  # For statistical data visualization
```

# Import dataset

The dataset is available at: https://www.kaggle.com/datasets/danielgrijalvas/movies/data

```python
# Read the CSV file 'movies.csv' into a DataFrame 'df'
df = pd.read_csv("movies.csv")

# Display the DataFrame 'df'
df
```

# Clean Dataset

```python
# Remove duplicate rows from the DataFrame 'df' based on the 'name' column
df.drop_duplicates(subset="name")
```

```python
# Remove the specified columns from the DataFrame 'df'
df.drop(columns=["writer", "star", "country", "votes", "runtime", "released"], axis=1, inplace=True)
```

```python
# Get the shape of the DataFrame 'df', which returns a tuple representing the dimensions (number of rows, number of columns)
df.shape
```

```python
# Display a concise summary of the DataFrame 'df', including the number of non-null entries and data types of each column
df.info()
```

```python
# Calculate the number of missing values in each column of the DataFrame 'df'
df.isna().sum()
```

```python
df.isna().sum() / len(df) * 100
```

```python
# Calculate the mode of the 'rating' column in the DataFrame 'df'
df.rating.mode()
```

```python
# Fill missing values in the 'rating' column with the mode value
df.fillna({"rating": "R"}, inplace=True)
```

```python
# Remove rows from the DataFrame 'df' that have missing values in the 'score' or 'company' columns
df.dropna(subset=["score", "company"], inplace=True)
```

```python
# Fill missing values in the 'gross' column with the mean value of the 'gross' column
df.fillna({"gross": df.gross.mean()}, inplace=True)
```

```python
# Fill missing values in the 'budget' column with the mean value of the 'budget' column
df.fillna({"budget": df.budget.mean()}, inplace=True)
```

```python
# We can notice that we no longer have any NaN data
df.isna().sum() / len(df) * 100
```

```python
# Generate descriptive statistics for the numerical columns in the DataFrame 'df'
df.describe()
```

```python
# Display the DataFrame 'df' after cleaning and preprocessing
df
```

# First Analysis

```python
# Group the DataFrame 'df' by the 'director' column, count the number of movies for each director,
# sort the result by the 'name' column in descending order, and select the top 10 directors
df_top_director = df.groupby("director").count().sort_values(by="name", ascending=False).head(10)
```

```python
# Visualize the top 10 directors by the number of movies
sns.barplot(data=df_top_director, x="name", y="director", hue="company")

# Display the plot
plt.show()
```

```python
# Group the DataFrame 'df' by the 'company' column, count the number of movies for each company,
# sort the result by the 'name' column in descending order, and select the top 10 companies
df_top_company = df.groupby("company").count().sort_values(by="name", ascending=False).head(10)
```

```python
# Visualize the top 10 companies by the number of movies
sns.barplot(data=df_top_company, x="name", y="company", hue="company")

# Display the plot
plt.show()
```

```python
# Group the DataFrame 'df' by the 'company' column, sum the 'budget' and 'gross' columns for each company,
# sort the result by the 'budget' column in descending order, and select the top 10 companies
df_budget_gross = df.groupby("company")[["budget", "gross"]].sum().sort_values(by="budget", ascending=False).head(10)
```

```python
# Visualize the top 10 companies by their gross earnings
sns.barplot(data=df_budget_gross, x="gross", y="company", hue="company")

# Display the plot
plt.show()
```

# Save the dataframe cleaned

```python
# Save the DataFrame 'df' to a CSV file named 'cleanMovies.csv' without the index column
df.to_csv("cleanMovies.csv", index=False)
```
